# Data Preparation

```python
# Preparing Data for training 
# Train
mnist_train_values = mnist_train.loc[:,"pixel0":]
mnist_train_label = mnist_train.loc[:,"label"]
```

```python
# transforming image pixel values from 0 to 255 to 0 to 1 by dividing the values by 255
mnist_train_values = mnist_train_values/255 # Train
mnist_test = mnist_test/255 # Test
```

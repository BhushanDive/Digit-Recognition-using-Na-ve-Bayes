# PCA

```python
from sklearn import decomposition
from sklearn.preprocessing import StandardScaler
```

```python
# Computing Standardised Data
standardised_scalar = StandardScaler()
standardised_data = standardised_scalar.fit_transform(mnist_train_values)
standardised_data.shape
```

```python
# Calculation of Convariance Matrix
cov_matrix = np.matmul(standardised_data.T,standardised_data)
cov_matrix.shape
```

```python
# Calculation of Eigen Vector and Values
from scipy.linalg import eigh
lambdas, vectors = eigh(cov_matrix,eigvals=(782,783))
vectors.shape
```

```python
vectors = vectors.T
vectors.shape
```

```python
#Calculating unit vectors u1 = v1 and new coordinates
new_coordinates = np.matmul(vectors,standardised_data.T)
print(new_coordinates.shape)
new_coordinates = np.vstack((new_coordinates,mnist_train_label)).T
```

```python
df_new = pd.DataFrame(new_coordinates, columns=["f1","f2","labels"])
df_new.head()
```

```python
# Plotting FacetGrid
sns.FacetGrid(df_new, hue = "labels", height = 6).map(plt.scatter, "f1", "f2").add_legend()
plt.show()
```

```python
from sklearn import decomposition

pca = decomposition.PCA()
pca.n_components = 2
pca_data = pca.fit_transform(standardised_data)
pca_data.shape
```

```python
pca_data = np.vstack((pca_data.T, mnist_train_label)).T
```

```python
df_PCA = pd.DataFrame(new_coordinates, columns=["f1", "f2", "labels"])
df_PCA.head()
```

# PCA Dimensionality Reduction

```python
pca.n_components = 784
pca_data = pca.fit_transform(standardised_data)
percent_variance_retained = pca.explained_variance_ / np.sum(pca.explained_variance_)

cum_variance_retained = np.cumsum(percent_variance_retained)
```

```python
plt.figure(1, figsize=(10, 6))
plt.clf()
plt.plot(cum_variance_retained, linewidth=2)
plt.axis("tight")
plt.grid()
plt.xlabel("number of compoments")
plt.ylabel("cumulative variance retained")
plt.savefig("pca_cumulative_variance.png")
plt.show()
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
print('success')
```

```python
naive_bayes_model = GaussianNB()
```

```python
fit_modelNB = naive_bayes_model.fit(x_train,y_train)
```

```python
predictions = fit_modelNB.predict(x_test)
conf_matrix = confusion_matrix(y_test, predictions)
print(conf_matrix)
```

```python
#Diagonal elements are true positives

# Taking sum of all TP 
def diagonal_ele_sum(conf_matrix):
    sum = 0
    for i in range(10):
        for j in range(10):
            if (i==j):
                sum += conf_matrix[i,j]
    return sum
```

```python
sum = diagonal_ele_sum(conf_matrix)
print(sum)
print(f'Accuracy %:{sum/8400}') # dividing by 8400 as it is the number of rows in test dataset.
# Finding Accuracy of the model
```

```python
predictions_2 = fit_modelNB.predict(x_train)
conf_matrix_2 = confusion_matrix(y_train, predictions_2)
print(conf_matrix_2)
```

```python
sum_2 = diagonal_ele_sum(conf_matrix_2)
print(sum_2)
print(f'Accuracy %: {sum_2/33600}')
```

```python
# Creating a heatmap of testing dataset

plt.imshow(conf_matrix, cmap='inferno',interpolation='nearest')
plt.xlabel('predictions')
plt.ylabel('actual_values')
plt.show()
```
